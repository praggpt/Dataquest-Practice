```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
women_degrees = pd.read_csv('percent-bachelors-degrees-women-usa.csv')
```

```python
plt.plot(women_degrees['Year'], women_degrees['Biology'])
```

```python
plt.show()
```

```python
plt.plot(women_degrees['Year'], women_degrees['Biology'], color='blue', label = 'Women')
plt.plot(women_degrees['Year'], 100-women_degrees['Biology'], color='green', label='Men')
plt.legend(loc='upper right')
plt.title('Percentage of Biology Degrees Awarded by Gender')
plt.show()
```

```python
plt.plot(women_degrees['Year'], 100-women_degrees['Biology'], color='green')
plt.title('Men')
plt.show()
```

## Removing non-data-ink


```python
fig, ax = plt.subplots()
plt.plot(women_degrees['Year'], women_degrees['Biology'], color='blue', label = 'Women')
plt.plot(women_degrees['Year'], 100-women_degrees['Biology'], color='green', label='Men')
plt.legend(loc='upper right')
plt.title('Percentage of Biology Degrees Awarded by Gender')
ax.tick_params(bottom='off', top='off', left='off', right='off')
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['top'].set_visible(False)
plt.show()
```

```python
major_cats = ['Biology','Computer Science','Engineering','Math and Statistics']
fig = plt.figure(figsize=(12,12))
for major in range(0,4):
    ax = fig.add_subplot(2,2,major+1)
    ax.plot(women_degrees['Year'],women_degrees[major_cats[major]], c = 'blue', label='Women')
    ax.plot(women_degrees['Year'], 100-women_degrees[major_cats[major]],c='green', label = 'Men')
    for key,spine in ax.spines.items():
        spine.set_visible(False)
    ax.set_title(major_cats[major])
    ax.tick_params(bottom='off', top='off', right='off', left='off')
    ax.set_xlim(1968,2011)
    ax.set_ylim(0,100)
    
    
plt.legend(loc='upper right')
plt.show()
```

# Color, layout and annotation

```python
blue = (0,107/255,164/255)
orange=(255/255,128/255,14/255)

major_cats = ['Biology','Computer Science','Engineering','Math and Statistics']
fig = plt.figure(figsize=(12,12))
for major in range(0,4):
    ax = fig.add_subplot(2,2,major+1)
    ax.plot(women_degrees['Year'],women_degrees[major_cats[major]], c = blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[major_cats[major]],c=orange, label = 'Men',linewidth=3)
    for key,spine in ax.spines.items():
        spine.set_visible(False)
    ax.set_title(major_cats[major])
    ax.tick_params(bottom='off', top='off', right='off', left='off')
    ax.set_xlim(1968,2011)
    ax.set_ylim(0,100)
    
    
plt.legend(loc='upper right')
plt.show()
```

## Arranging charts in one line
To see more clearly and ordering in descending order of gender ratio gap 

```python
blue = (0,107/255,164/255)
orange=(255/255,128/255,14/255)

stem_cats = ['Engineering', 'Computer Science', 'Psychology', 'Biology', 'Physical Sciences', 'Math and Statistics']
fig = plt.figure(figsize=(18,3))
for major in range(0,6):
    ax = fig.add_subplot(1,6,major+1)
    ax.plot(women_degrees['Year'],women_degrees[stem_cats[major]], c = blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[major]],c=orange, label = 'Men',linewidth=3)
    for key,spine in ax.spines.items():
        spine.set_visible(False)
    ax.set_title(stem_cats[major])
    ax.tick_params(bottom='off', top='off', right='off', left='off')
    ax.set_xlim(1968,2011)
    ax.set_ylim(0,100)
    
    
plt.legend(loc='upper right')
plt.show()
```

# Guided project


```python
%matplotlib inline
```

```python
stem_cats = ['Psychology', 'Biology', 'Math and Statistics', 'Physical Sciences', 'Computer Science', 'Engineering']
lib_arts_cats = ['Foreign Languages', 'English', 'Communications and Journalism', 'Art and Performance', 'Social Sciences and History']
other_cats = ['Health Professions', 'Public Administration', 'Education', 'Agriculture','Business', 'Architecture']
```

```python
fig = plt.figure(figsize=(16,20))

for major in range(0,6):
    ax = fig.add_subplot(6,3,major*3+1)
    ax.plot(women_degrees['Year'],women_degrees[stem_cats[major]], linewidth=3, c=blue)
    ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[major]], linewidth=3, c=orange)
    ax.set_xlim(1968,2011)
    ax.set_ylim(0,100)
    if major == 0:
        ax.text(2005,87,'Women')
        ax.text(2005,15,'Men')
    if major == 5:
        ax.text(2005, 85, 'Men')
        ax.text(2005, 5, 'Women')
    ax.set_title(stem_cats[major])
    for key, spine in ax.spines.items():
        spine.set_visible(False)
    ax.tick_params(bottom='off', top='off', left='off', right ='off')

plt.show()
```

```python
midline = (171/255,171/255,171/255)

fig = plt.figure(figsize=(16,20))
# for stem majors
for major in range(0,6):
    ax = fig.add_subplot(6,3,major*3+1)
    ax.plot(women_degrees['Year'],women_degrees[stem_cats[major]], linewidth=3, c=blue)
    ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[major]], linewidth=3, c=orange)
    ax.set_xlim(1968,2011)
    ax.set_ylim(0,100)   
    ax.set_yticks([0,100]) # removing clutter to show only 0 and 100
    ax.axhline(50, c=midline, alpha=0.3) # horizontal line to show where the 50:50 mark is
    ax.tick_params(bottom='off', top='off', left='off', right ='off',labelbottom='off')
    if major == 0:
        ax.text(2005,87,'Women')
        ax.text(2005,15,'Men')
    if major == 5:
        ax.text(2005, 85, 'Men')
        ax.text(2005, 5, 'Women')
        ax.tick_params(labelbottom='on')
    ax.set_title(stem_cats[major])
    for key, spine in ax.spines.items():
        spine.set_visible(False)

#for lib-arts majors
for major in range(0,5):
    ax=fig.add_subplot(6,3,major*3+2)
    ax.plot(women_degrees['Year'],women_degrees[lib_arts_cats[major]], linewidth=3, c=blue)
    ax.plot(women_degrees['Year'],100-women_degrees[lib_arts_cats[major]], linewidth=3,c=orange)
    ax.set_ylim(0,100)
    ax.set_xlim(1968,2011)
    ax.set_title(lib_arts_cats[major])
    ax.set_yticks([0,100]) # removing clutter to show only 0 and 100
    ax.axhline(50, c=midline, alpha=0.3) # horizontal line to show where the 50:50 mark is
    for key,spine in ax.spines.items():
        spine.set_visible(False)
    ax.tick_params(bottom='off', top='off', left='off', right ='off',labelbottom='off')
    if major == 0:
        ax.text(2005,83,'Women')
        ax.text(2005,15,'Men')
    if major == 4:
        ax.tick_params(labelbottom='on')


#for other_cats majors
for major in range(0,6):
    ax=fig.add_subplot(6,3,major*3+3)
    ax.plot(women_degrees['Year'],women_degrees[other_cats[major]], linewidth=3, c=blue)
    ax.plot(women_degrees['Year'],100-women_degrees[other_cats[major]], linewidth=3,c=orange)
    ax.set_ylim(0,100)
    ax.set_xlim(1968,2011)
    ax.set_title(other_cats[major])
    ax.set_yticks([0,100]) # removing clutter to show only 0 and 100
    ax.axhline(50, c=midline, alpha=0.3) # horizontal line to show where the 50:50 mark is
    for key,spine in ax.spines.items():
        spine.set_visible(False)
    ax.tick_params(bottom='off', top='off', left='off', right ='off', labelbottom='off')
    if major == 0:
        ax.text(2005,77,'Women')
        ax.text(2005,18,'Men')  
    if major == 5:
        ax.text(2005, 65, 'Men')
        ax.text(2005, 35, 'Women')
        ax.tick_params(labelbottom='on')

plt.savefig('gender_degrees.png')
plt.show()
```
