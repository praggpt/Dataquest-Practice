```python
import pandas as pd
```

```python
titanic = pd.read_csv('train.csv')
```

```python
titanic=titanic[['Survived','Pclass','Sex','Age','SibSp','Parch','Fare','Embarked']]
```

```python
titanic.head()
```

```python
titanic=titanic.dropna()
```

```python
titanic.describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.distplot(titanic['Fare'])
plt.show()
```

```python
sns.distplot(titanic['Age'])
plt.show()
```

```python
sns.kdeplot(titanic['Age'], shade=True)
plt.xlabel('Age')
plt.show()
```

```python
sns.set_style('white')
```

```python
sns.kdeplot(titanic['Age'], shade=True)
sns.despine(left=True, bottom=True)
plt.xlabel('Age')
plt.show()
```

# Creating multiple plots for distinct values in a column

```python
x = sns.FacetGrid(titanic, col= 'Pclass', size=6 )
x.map(sns.kdeplot, 'Age',shade=True)
sns.despine(bottom=True, left=True)
plt.show()
```

```python
g = sns.FacetGrid(titanic, col="Survived", row="Pclass")
g.map(sns.kdeplot, "Age", shade=True)
sns.despine(left=True, bottom=True)
plt.show()
```

```python
p=sns.FacetGrid(titanic, col='Survived', row= 'Pclass', hue='Sex', size=3)
p.map(sns.kdeplot, 'Age', shade=True).add_legend()
sns.despine(left=True, bottom=True)

plt.show()
```
